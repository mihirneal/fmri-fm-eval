```python
from lightning_sdk import Studio, Machine, Job
import numpy as np
```

```python
studio = Studio(name="connor-preprocess", teamspace="medarc", org="medarc")
```

```python
num_jobs = 715
batch_size = 54
num_batches = num_jobs // batch_size

offsets = np.round(np.linspace(0, num_jobs, num_batches + 1)).astype(np.int64)
starts = offsets[:-1]
stops = offsets[1:]
batch_sizes = stops - starts
for ii, (start, stop) in enumerate(zip(starts, stops)):
    print(ii, start, stop, stop - start)
print(batch_sizes.sum())
```

```python
batch_ids = range(num_batches)

for batch_id in batch_ids:
    start = starts[batch_id]
    stop = stops[batch_id]
    name = f"ppmi_fmriprep_v0_{batch_id:02d}"

    job = Job.run(
        command=f"bash fmri-fm-eval/datasets/PPMI/scripts/preprocess_batch.sh {start} {stop}",
        name=name,
        machine=Machine.DATA_PREP_MAX,
        studio=studio,
        interruptible=False,
    )
    print(job.name)
```
